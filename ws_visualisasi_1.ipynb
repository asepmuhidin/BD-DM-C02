```python
data_file='/mnt/MyDrive/DMBDUPB/dataset/Global_Superstore2.xlsx'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel(data_file)
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data['Country'].value_counts()
```

```python
data['Country'].unique()
```

```python
market_ina=data[data.Country=='Indonesia']
market_ina.head()
```

```python
sales_state=market_ina.groupby('State')['Sales'].sum().sort_values(ascending=False)
sales_state
```

```python
plt.xticks(rotation=90)
plt.title('Penjualan di setiap Provinsi Indonesia')
plt.xlabel('Provinsi')
plt.ylabel('Penjualan')
plt.bar(sales_state.index,sales_state);
```

```python
sales_state_asc=market_ina.groupby('State')['Sales'].sum().sort_values(ascending=True)
plt.xticks(rotation=90)
plt.title('Penjualan di setiap Provinsi Indonesia')
plt.ylabel('Provinsi')
plt.xlabel('Penjualan')
plt.barh(sales_state_asc.index,sales_state_asc);
```

```python
plt.pie(sales_state.head(),labels=sales_state.head().index,autopct='%1.1f%%');
```
