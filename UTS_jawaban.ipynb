```python
import pandas as pd

# Data sample
data = {
    "ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "EmployeeName": [
        "Alice Smith", "Bob Johnson", "Charlie Lee", "Diana King", "Evan Wright",
        "Fiona Adams", "George Young", "Hannah Baker", "Ian Scott", "Julia Clark"
    ],
    "Title": [
        "Software Engineer", "Data Analyst", "Project Manager", "HR Specialist", "Marketing Lead",
        "UX Designer", "DevOps Engineer", "Accountant", "Business Analyst", "Legal Advisor"
    ],
    "Dept": [
        "IT", "PMO", "PMO", "HR", "HR",
        "HR", "IT", "PMO", "IT", "IT"
    ],
    "Salary": [75000, 68000, 82000, 60000, 71000, 69000, 77000, 65000, 73000, 80000],
    "Status": ["Full-Time", "Full-Time", "Full-Time", "Part-Time", "Full-Time",
               "Part-Time", "Full-Time", "Full-Time", "Full-Time", "Part-Time"]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Menampilkan DataFrame
df.head(10)
```

```python
df.shape
```

```python
print(f'Jumlah data karyawan adalah {df.shape[0]}')
```

```python
print(f'Rata-rata Gaji  karyawan adalah {df.Salary.mean()}')
```

```python
df[df.Salary==df.Salary.min()]
```

```python
EmployeeOfMinSalary=df[df.Salary==df.Salary.min()]['EmployeeName']
EmployeeOfMaxSalary=df[df.Salary==df.Salary.max()]['EmployeeName']
```

```python
print(f'Gaji Minimum karyawan adalah {df.Salary.min()} bernama {EmployeeOfMinSalar.values[0]}')
print(f'Gaji Maksimum karyawan adalah {df.Salary.max()} bernama {EmployeeOfMaxSalary.values[0]}')
```
